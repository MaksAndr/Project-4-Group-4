```python
df_cleaned = df.drop(axis='columns', columns=['Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_1','Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_2'])
```

```python
df_cleaned=df_cleaned.rename(columns={'Attrition_Flag':'Churned', 'Customer_Age':'Age'})

# df_cleaned = df_cleaned.replace(to_replace="Existing Customer", value=0)
# df_cleaned = df_cleaned.replace(to_replace="Attrited Customer", value=1)

df_cleaned = df_cleaned.replace(to_replace="M", value=0)
df_cleaned = df_cleaned.replace(to_replace="F", value=1)
```

```python
df_cleaned = df_cleaned.replace(to_replace='High School', value=0)
df_cleaned = df_cleaned.replace(to_replace='College', value=1)
df_cleaned = df_cleaned.replace(to_replace='Graduate', value=2)
df_cleaned = df_cleaned.replace(to_replace='Post-Graduate', value=3)
df_cleaned = df_cleaned.replace(to_replace='Doctorate', value=4)
df_cleaned = df_cleaned.replace(to_replace='Uneducated', value=5)
df_cleaned = df_cleaned.replace(to_replace='Unknown', value=6)


```

```python
df_cleaned = df_cleaned.replace(to_replace="Less than $40K", value=0)
df_cleaned = df_cleaned.replace(to_replace="$40K - $60K", value=1)
df_cleaned = df_cleaned.replace(to_replace="$60K - $80K", value=2)
df_cleaned = df_cleaned.replace(to_replace="$80K - $120K", value=3)
df_cleaned = df_cleaned.replace(to_replace="$120K +", value=4)
df_cleaned = df_cleaned.replace(to_replace="Unknown", value=5)
```

```python
df_cleaned = df_cleaned.replace(to_replace="Single", value=0)
df_cleaned = df_cleaned.replace(to_replace="Married", value=1)
df_cleaned = df_cleaned.replace(to_replace="Unknown", value=2)
df_cleaned = df_cleaned.replace(to_replace="Divorced", value=3)
```

```python
df_cleaned = df_cleaned.replace(to_replace="Blue", value=0)
df_cleaned = df_cleaned.replace(to_replace="Silver", value=1)
df_cleaned = df_cleaned.replace(to_replace="Gold", value=2)
df_cleaned = df_cleaned.replace(to_replace="Platinum", value=3)
```

```python
df_scaled=StandardScaler().fit_transform(df_cleaned[["Age","Gender","Education_Level","Marital_Status","Income_Category","Card_Category","Credit_Limit"]])

df_transformed = pd.DataFrame(df_scaled, columns=["Age","Gender","Education_Level","Marital_Status","Income_Category","Card_Category","Credit_Limit"])


df_transformed.head()
```
